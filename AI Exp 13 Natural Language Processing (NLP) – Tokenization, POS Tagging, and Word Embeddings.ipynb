

Task 1 & 2: Tokenization and POS (Parts of Speach) Tagging using NLTK



```python
# Install and import necessary libraries
# Ensure nltk is available in the environment
!pip install nltk --quiet

import nltk
import os

# Set the NLTK data path explicitly
nltk.data.path.append('/root/nltk_data')

# Download necessary resources quietly to avoid excessive output
nltk.download('punkt')  # Tokenizer models
nltk.download('averaged_perceptron_tagger')  # POS tagger
nltk.download('universal_tagset')  # Simplified POS tags
nltk.download('punkt_tab')

# Import tokenization and tagging tools
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag

# Real-life Sample Text (could be part of an AI news article)
text = "Artificial Intelligence is transforming the world. Machines now understand human language like never before. Companies such as Google and OpenAI are investing heavily in AI research."
```

```python
# Sentence Tokenization: Break the paragraph into individual sentences
sentences = sent_tokenize(text)
print("ðŸ”¹ Sentence Tokenization:")
for i, sentence in enumerate(sentences, 1):
    print(f"{i}. {sentence}")

# Word Tokenization: Split the entire paragraph into words and punctuation
words = word_tokenize(text)
print("\nðŸ”¹ Word Tokenization:")
print(words)

# POS Tagging: Assign grammatical tags to each word (noun, verb, etc.)
# We use the 'universal' tagset for simplified and more interpretable tags
pos_tags = pos_tag(words, tagset='universal')
print("\nðŸ”¹ Part-of-Speech Tagging:")
for word, tag in pos_tags:
    print(f"{word} -> {tag}")

# Example of POS tags:
# 'Artificial' -> ADJ (Adjective)
# 'Machines' -> NOUN
# 'understand' -> VERB
```

Task 3: Word Embeddings using spacy



```python
# âœ… Install spaCy and download the English model
!pip install -U spacy --quiet
!python -m spacy download en_core_web_md

# âœ… Load the spaCy model
import spacy
nlp = spacy.load("en_core_web_md")

# âœ… Input sentences
sentences = [
    "The king and the queen rule the kingdom.",
    "A computer can solve complex problems.",
    "The government passed a new law."
]

# âœ… Analyze word vectors
doc = nlp(" ".join(sentences))

# âœ… Access vector for a word
print("\nðŸ”¹ Vector for 'king':")
print(nlp("king").vector[:10])  # First 10 values

# âœ… Word similarity
print("\nðŸ”¹ Similarity between 'king' and 'queen':")
print(nlp("king").similarity(nlp("queen")))

print("\nðŸ”¹ Similarity between 'government' and 'law':")
print(nlp("government").similarity(nlp("law")))

# âœ… Find most similar in context manually
print("\nðŸ”¹ Contextual similarity in a sentence:")
token_list = [token.text for token in doc if token.has_vector and not token.is_stop]
for token in token_list:
    print(f"Similarity of '{token}' to 'government': {nlp(token).similarity(nlp('government')):.4f}")
```

```python
# âœ… Install spaCy and download the English medium-sized model
!pip install -U spacy --quiet
!python -m spacy download en_core_web_md

# âœ… Load the medium model which has vectors
import spacy
nlp = spacy.load("en_core_web_md")

# âœ… Vocabulary words to compare
words = ["king", "queen", "man", "woman", "government", "law", "computer", "technology", "president", "kingdom"]

# âœ… 1. Word similarity between two known terms
print("\nðŸ”¹ Similarity between 'king' and 'queen':")
print(nlp("king").similarity(nlp("queen")))

# âœ… 2. Vector representation of a word
print("\nðŸ”¹ Vector for word 'computer':")
print(nlp("computer").vector[:10])  # Display first 10 dimensions

# âœ… 3. Most similar words to 'government' based on cosine similarity
target = nlp("government")
similarities = []
for word in words:
    if word != "government":
        similarity = target.similarity(nlp(word))
        similarities.append((word, similarity))

# âœ… Sort and print top 5 most similar words
sorted_similarities = sorted(similarities, key=lambda x: x[1], reverse=True)[:5]
print("\nðŸ”¹ Most similar words to 'government':")
for word, score in sorted_similarities:
    print(f"{word}: {score:.4f}")
```
