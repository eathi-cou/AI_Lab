```python
# ----------------------------------------
# 2. NumPy Basics
# ----------------------------------------
import numpy as np

# Create a 2D array (Matrix)
A = np.array([[1, 2], [3, 4]])
print("\nMatrix A:")
print(A)

# Transpose of matrix
print("\nTranspose of A:")
print(A.T)

# Matrix multiplication
B = np.array([[5, 6], [7, 8]])
print("\nMatrix B:")
print(B)

product = np.dot(A, B)
print("\nMatrix Product A x B:")
print(product)
```

```python
# ----------------------------------------
# 3. Pandas Basics
# ----------------------------------------
import pandas as pd

# Sample student score data
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [80, 70, 60],
    'Physics': [85, 78, 55],
    'Chemistry': [90, 75, 65]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("\nStudent Scores Table:")
print(df)

# Access column and compute average
print("\nAverage Marks in Each Subject:")
print(df.mean(numeric_only=True))

# Add a new column with average score per student
df['Average'] = df[['Math', 'Physics', 'Chemistry']].mean(axis=1)
print("\nDataFrame with Average Score:")
print(df)

# Filter students scoring above 75 average
print("\nStudents scoring above 75:")
print(df[df['Average'] > 75])
```

```python
# ----------------------------------------
# 4. Data Visualization using Matplotlib
# ----------------------------------------
import matplotlib.pyplot as plt

# Bar chart for Math scores
plt.figure(figsize=(6, 4))
plt.bar(df['Name'], df['Math'], color='skyblue')
plt.title("Math Scores of Students")
plt.xlabel("Student")
plt.ylabel("Math Score")
plt.grid(axis='y')
plt.show()

# Line plot for all subjects
plt.figure(figsize=(8, 5))
plt.plot(df['Name'], df['Math'], marker='o', label='Math')
plt.plot(df['Name'], df['Physics'], marker='x', label='Physics')
plt.plot(df['Name'], df['Chemistry'], marker='s', label='Chemistry')
plt.title("Subject-wise Performance")
plt.xlabel("Student")
plt.ylabel("Scores")
plt.legend()
plt.grid(True)
plt.show()
```
